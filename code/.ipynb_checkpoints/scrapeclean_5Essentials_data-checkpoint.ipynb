# Scrape the 5 Essentials Data

```python
#trying to scrape 5-essentials data

import requests
import pandas as pd
from selenium import webdriver
import os

import numpy as np
import pickle
import seaborn as sns

chromedriver = "/Applications/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
#First import a list of schools in CPS with basic information 
#about high schools - unique SchoolID, name, network, SQRP points and 
#rating, and type (combo, elementary, or high school)
school_list = pd.read_csv('../data/schoolsdata.csv')

#Make a really small set of schools to test out web scraping loop
school_list2=school_list.head(3)
```

## Scrape the 5E data here

```python
#The URL for getting 5Essentials data for each school follows a format where
#the second to last directory matches the unique SchoolID
base_url = "https://5-essentials.org/cps/5e/2018/s/{}/essentials/"

#Dictionary will hold the 5Essentials data associated with each SchoolID key
df_dict = {}

#Loop over every school in the list and plug the schoolid into the URL
for schoolid in school_list2['SchoolID']:
    url = base_url.format(schoolid)
    driver = webdriver.Chrome(chromedriver)
    
    try:
        #Open and grab the source code for each page 
        #(have to use Selenium to deal with Javascript)
        driver.get(url)
        temp=driver.page_source
        
        #Grab just a table of the 5Essentials data and convert it to a row of 
        #data where each column heading is one of the 5Essentials metrics, save to dict
        tempdf = (pd.read_html(temp))[0].transpose()
        tempdf = tempdf.rename(columns=tempdf.iloc[0])
        tempdf = tempdf.sort_index(axis=1)
        df_dict[schoolid]=tempdf.iloc[1]
        driver.close()
    except ValueError:
        #Some schools have closed or consolidated in the last 5Essentials measure year
        #and have no webpage
        df_dict[schoolid] = 'No data'
        driver.close()
```

```python
# Convert the data into a dataframe and make it wide, then get the columns
school_df = pd.DataFrame.from_dict(df_dict)
school_df =school_df.transpose()
columns_list = school_df.columns
```

## Clean the 5E data up a little bit 

```python
#The score is stored in the format ['30 weak'] - to isolate the number,
#split on a space and leave just the first entry in the resulting list, which should be
#numeric
for category in columns_list:
    school_df[category] = school_df[category].str.split(' ',1).str.get(0)
    
```

# Bring in other sources of data

## Charter Unionization Data

This was created by me by hand - basically I made a spreadsheet where I entered which charters were unionized, and what date their first contract went into effect. This was determined through a combination of news articles, the CTU website, and actual PDFs of the contract found through Googling.

```python
# Some charters are unionized or not - I wanted to include data on unionized/not 
# as well as length of time a charter had been under a union contract
union_info= pd.read_csv('../data/Chicago_charter_union.csv')

#Convert date of first contract to datetime object, then make a column for the #
# of days under contract that elapsed until 5Essentials survey collection
union_info['YEAR_CONT']=pd.to_datetime(union_info['YEAR_CONT'])
union_info =union_info.drop(['SchoolName','Type'],axis=1)
datetimeSurv=pd.to_datetime("03/01/19")
union_info['Days_Cont']=(datetimeSurv-union_info['YEAR_CONT']).dt.days
```

```python
# Join the union data with the general schools and 5Es data
with_union = tempdf.merge(union_info,on='SchoolID',how='left')
```

## CPS Demographic Data

This data is organized by SchoolID, so it merges really easily with our 5Es data and list of school names

```python
# Also upload CPS demographic data, which contains population size, %bilingual, %sped, 
# %FRLunch, and merge with the data so far
demographics= pd.read_csv('../data/CPS_demographics2.csv')
demographics=demographics.drop(['School Name'],axis=1)
fullschools = with_union.merge(demographics,on='SchoolID',how='left')

# Make the unique schoolID the index
fullschools.index = fullschools['SchoolID']
```

## ISBE Data

ISBE data is not organized by school ID, but by ISBE's own identification system (called the RTSC). I had to find another data table that paired up the RTSC with the SchoolID for each school, and merge those, before I could integrate my ISBE data with my scraped data.

```python
# Let's also bring in some Illinois State Board of Ed data, which includes a lot of 
# useful data but can be incomplete because charters don't usually report data to ISBE


ISBE_data = pd.read_csv('../data/Report-Card-Public-Data-Set.csv', encoding='latin-1')
ISBE_data= ISBE_data.sort_values('School Name')
ISBE_data = ISBE_data.reset_index()
ISBE_data=ISBE_data.drop('index',axis=1)

# ISBE data is 382 columns wide - lets just grab the ones that seem useful and are
# generally populated with values for CPS schools
# also rename the weird column names
ISBE_data_smaller = ISBE_data[['RCDTS','Student Attendance Rate','Student Mobility Rate','Teacher Retention Rate','Avg Class Size Ð All Grades','Principal Turnover within 6 Years']]
ISBE_data_smaller.rename(columns={'Avg Class Size Ð All Grades':'AvgClassSize','RCDTS':'ISBE ID'},inplace=True)
```

```python
ISBE_data_smaller
```

```python
# A helpful person created a CSV file with all the different synonyms for every CPS school
# plus a lot other information
CPS_names = pd.read_csv('../data/CPS_Names.csv')

# SchoolIDs got converted to floats, so fix that
CPS_names['SchoolID']=CPS_names['SchoolID'].fillna(0).astype(int)

# Isolate the columns that match SchoolID up with ISBE ID - this will allow us to merge
# ISBE data with CPS data. Keep location (lat,long) because maybe we can do something 
# with geography
CPS_names_fewer=CPS_names[['SchoolID','ISBE ID','Location']]

# Merge the ISBE data together with the CPS schoolID
names_ISBE = CPS_names_fewer.merge(ISBE_data_smaller,on='ISBE ID',how='left')

# Now we can merge the ISBE data with CPS, using schoolID to identify unique CPS schools
newfull = school5Es_df.merge(names_ISBE,on='SchoolID',how='left')

# Set the schoolID to be the index again
newfull.index = newfull['SchoolID']
```

```python
# Pickle this! Our data is ready to begin modeling
with open('schools_updatedData', 'wb') as to_write:
    pickle.dump(newfull, to_write)
```
